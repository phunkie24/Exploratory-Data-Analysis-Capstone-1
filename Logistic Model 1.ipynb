```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Logit Function

```python
def logit(x):
    return np.log(x/(1-x))
x=np.arange(0.001, 0.99, 0.0001)
y=[logit(n) for n in x]

plt.plot(x,y)
plt.xlabel("Probability")
plt.ylabel("Logit - L")
plt.show()
```

# Simoid Function

```python
def sigmoid (x):
    return (1/(1+np.exp(-x)))
x=np.arange(-10, 10, 0.0001)
y=[sigmoid(n) for n in x]

plt.plot(x,y)
plt.xlabel("Logit - L")
plt.ylabel("Probabibility")
plt.show()
    
```
